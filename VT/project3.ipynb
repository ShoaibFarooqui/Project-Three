```python
from datetime import datetime as dt
import pandas as pd
import numpy as np
import os
```

```python
#Read in original csv
aviation_df = pd.read_csv("C:\\Users\\Junior\\Desktop\\Project3\\NotCurrentlyUsed\\AviationData.csv", encoding='ISO 8859-1')
aviation_df.columns
#aviation_df.head()
```

```python
#Renamed columns
rename_aviation = aviation_df[{'Event.Date', 'Location', 'Country', 'Latitude', 'Longitude', 'Airport.Name', 'Air.carrier', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured', 'Weather.Condition'}]
rename_aviation.rename(columns = {'Event.Date': 'Event_Date', 'Airport.Name':'Airport_Name','Air.carrier':'Air_Carrier','Total.Fatal.Injuries': 'Total_Fatal_Injuries',
 'Total.Serious.Injuries':'Total_Serious_Injuries','Total.Minor.Injuries': 'Total_Minor_Injuries', 'Total.Uninjured':'Total_Uninjured', 'Weather.Condition':'Weather_Condition'}, inplace= True)
#cleaned_aviation
#Reordering columns
cleaned_aviation = rename_aviation[['Air_Carrier', 'Event_Date', 'Country', 'Location', 'Airport_Name','Latitude', 'Longitude', 'Weather_Condition', 'Total_Fatal_Injuries', 'Total_Serious_Injuries', 'Total_Minor_Injuries', 'Total_Uninjured']]
cleaned_aviation
```

```python
#Date Range and unique locations 2010-2020 for larger dataset
date_df = cleaned_aviation.loc[(cleaned_aviation['Event_Date'] >= '2010-01-01') & (cleaned_aviation['Event_Date'] < '2020-01-01')]
#City and State columns
date_df[['City', 'State']] = date_df['Location'].str.split(',', 1, expand = True)
date_df

```

```python
#Cleaned DataFrame
dropped_df = date_df.dropna()
dropped_df

```

```python
#Groupby to find total sums of injuries by state or country
#dropped_df.groupby(by=["State"])["Total_Fatal_Injuries"].sum()
country_fatal = dropped_df.groupby(by=["Country"])["Total_Fatal_Injuries"].sum()
country_uninjured = dropped_df.groupby(by=["Country"])["Total_Uninjured"].sum()
fatal_df = pd.DataFrame(country_fatal)
#fatal_df
uninjured_df = pd.DataFrame(country_uninjured)
#uninjured_df


```

```python
#merging country dataframes
merged_df = fatal_df.merge(uninjured_df, on='Country', how='right')
merged_df
```

```python
#State values
state_fatal = dropped_df.groupby(by=["State"])["Total_Fatal_Injuries"].sum()
state_f = pd.DataFrame(state_fatal)
state_uninjured = dropped_df.groupby(by=["State"])["Total_Uninjured"].sum()
state_u = pd.DataFrame(state_uninjured)
merged_state = state_f.merge(state_u, on='State', how='right')
merged_state
```

```python
#Exporting cleaned CSV
dropped_df.to_csv("cleaned_project3.csv", encoding = 'utf-8', index = False)
merged_df.to_csv("merged.csv", encoding = 'utf-8')
merged_state.to_csv("merged_state.csv", encoding = 'utf-8')
```
